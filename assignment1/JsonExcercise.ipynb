```python
import pandas as pd
import json
import numpy as np 
from pandas.io.json import json_normalize
#Find the top 10 major project themes (using column 'mjtheme_namecode')
# load json as string
data=json.load((open('data/world_bank_projects.json')))

# use normalization to create tables from nested element
project_themes=json_normalize(data, 'mjtheme_namecode')

#find frequency of each code value
frequency=project_themes.code.value_counts().sort_index().to_frame(name='count')
frequency['code']=frequency.index
frequency

#get distinct code, name pairs
project_themes=project_themes.replace('', np.nan)
project_themes=project_themes.dropna()
project_themes=project_themes.drop_duplicates().sort_values(by='code')

project_themes

#merge code with project_themes to get project name with the frequency
frequency.merge(project_themes).sort_values(by='count',ascending=False).head(10)

#In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.

# use normalization to create tables from nested element
missing_data=json_normalize(data, 'mjtheme_namecode')

#replace missing value by NaN
missing_data=missing_data.replace('', np.nan).sort_values(by='code')

#sort data by code and fill NaN 
#how to choose bfill or ffill?? can create problems at boundary
missing_data=missing_data.fillna(method='bfill').sort_index()
missing_data
```
